## y2=x+2

## y3=x+3

## y3=x+4

```python
x=np.arange(0,12,1)
y1=x+1
y2=x+2
y3=x+3
y4=x+4
plt.subplot(1, 2, 1)
plt.plot(t,y1, 'r--',color='blue', marker='*')
plt.plot(t,y2, 'go-',color='red',marker='.')
plt.ylabel('sun')
plt.xlabel('nus')
plt.title('sunun')
plt.legend('sn')
plt.subplot(1, 2, 2)
plt.plot(t,y3, 'r--',color='green', marker='+')
plt.plot(t,y4, 'go-',color='yellow',marker=',')
plt.ylabel('start')
plt.xlabel('end')
plt.title('startend')
plt.legend('se')
```
