```python
import matplotlib.pyplot as plt
a=10, 20, 30, 40
b='kor', 'emg','math','info' 
plt.bar(b,a, color = 'red')
plt.plot(b,a,'r--')
plt.ylabel('jumsu')
plt.xlabel('subject')
plt.title('Score')
plt.legend('value')
```
